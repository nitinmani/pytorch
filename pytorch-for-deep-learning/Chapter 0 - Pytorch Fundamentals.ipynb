Why random tensors?

Start with random nums --> look at data --> update random nums --> look at data --> update random numbers

### Matrix Multiplication

Two main ways of performing multiplication in neural network and deep learning
- element-wise multiplication
- matrix multiplication (dot product)

For matrix multiplication, the inner dimensions must match. Also, the resulting matrix has the shape of the outer dimensions (basic Lin Alg)

```python
print(tensor, "*", tensor)
print(tensor * tensor) # element-wise multiplication
```

```python
# Matrix Multiplication
torch.matmul(tensor, tensor) # 1 * 1 + 2 * 2 + 3 * 3
```

```python
%%time
value = 0
for i in range(len(tensor)):
    value += tensor[i] * tensor[i]
print(value)
```

```python
%%time
torch.matmul(tensor, tensor) # much faster than writing a for loop
```

### Reshaping, Stacking, Squeezing, and Unsqueezing

* Reshaping - Reshapes an input tensor to a defined shape
* View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking - combine multiple tensors on top of each other (vstack) or side by side
* Squeeze - Removes all `1` dimensions from a tensor
* Unsqueeze - Add a `1` dimension to a target tensor
* Permute - Return a view of the input with dimensions permuted (swapped) in a certain way

```python
x = torch.arange(1., 10.)
x, x.shape
```

```python
# add an extra dimension
x_reshaped = x.reshape(1, 7) # trying to squeeze tensor of 9 elements into 7
```

```python
x_reshaped = x.reshape(1, 9)
```

```python
x_reshaped = x.reshape(2, 9) # cannot double the amount of elements
```

```python
x_reshaped = x.reshape(9, 1)
x_reshaped, x_reshaped.shape # reshape must be compatible with the original size
```

```python
x_reshaped = x.reshape(3, 3)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1, 9)
z, z.shape
```

```python
# changing z changes x because a view of a tensor shares the same memory as the original tensor
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked, x_stacked.shape
```

```python
x_stacked = torch.stack([x, x, x, x], dim=1)
x_stacked, x_stacked.shape
```

```python
x_v_stacked = torch.vstack([x,x,x,x])
x_v_stacked, x_v_stacked.shape
```

```python
x_h_stacked = torch.hstack([x,x,x,x])
x_h_stacked, x_h_stacked.shape
```

```python
# torch.permute - rearranges the dimensions of the target tensor in a specified order
x_original = torch.rand(size=(224, 224, 3)) # [height, width, color_channels]

# permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0 to 1, 1 to 2, 2 to 0
print(f"Previous shape: {x_original.shape}")
print(f"New Shape: {x_permuted.shape}") # [color_channels, height, width]
```

### Reproducibility (Taking the Random out of Random)

`The NN learns: start with random numbers --> tensor operations --> update random numbers to try and make them better representations of the data --> loop`

To reduce the randomness in NNs and PyTorch is the concept of **random seed**

This flavors the randomness

```python
# create two random tensors
random_tensor_a = torch.rand(3,4)
random_tensor_b = torch.rand(3,4)

print(random_tensor_a)
print(random_tensor_b)
print(random_tensor_a == random_tensor_b)
```

```python
# random but reproducible tensors
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)

random_tensor_c = torch.rand(3,4)
random_tensor_d = torch.rand(3,4)

print(random_tensor_c)
print(random_tensor_d)
print(random_tensor_c == random_tensor_d)
```

```python
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)

random_tensor_c = torch.rand(3,4)

torch.manual_seed(RANDOM_SEED) # need to call each time to set the flavor
random_tensor_d = torch.rand(3,4)

print(random_tensor_c)
print(random_tensor_d)
print(random_tensor_c == random_tensor_d)
```
