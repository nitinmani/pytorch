Machine learning is turning things (data) into numbers and finding patterns in those numbers

Computer finds these pattersn using code and math

Traditional Programming --> you'll have some inputs, then write down some rules (program). This yields an output

Machine Learning --> you'll have some inputs (features) and desired outputs (labels), then the algorithm figures out the rules to determine the patterns between inputs (features) and outputs (labels). This specifically is SUPERVISED LEARNING

**Why even use Machine Learning or Deep Learning?**

For a complex problem, can you think of all of the rules?

ex) Driving a car involves a lot of rules. You probably learned this in a day but to list out all of the rules is probably not happening. ML and Deep Learning can help with this

If you can build a simple rule-based system that does not rquire ML, do that

**What is Deep Learning Good For?**

Problems with long lists of rules --> when the traditional approach fails, machine learning/deep learning may help (self driving cars)

Continually changing environments --> deep learning can adapt to new scenarios

Discovering insights within large collections of data --> hand-crafting rules for 101 different kinds of foods would be painful



What Deep Learning is NOT Good For

When you need explainabiity --> the patterns learned by a deep learning model are typically uninterpretable by a human

When the traditional approach is a better options --> if you can accomplish what you need with a simple rule-based system

When errors are unacceptable --> since the outputs of deep learning models aren't always predictable

When you don't have a lot of data --> deep learning models usually require a fairly large amount of data to produce great results

Traditional ML algorithms should be used on structured data (rows and columns)
Algorithm: Gradient boosted machine (XGBoost)
Random Forest, Gradient boosted models, Naive Bayes, Nearest Neighbor, SVM

Deep Learning algorithms are better on unstructured data. Data is not in a nice, standardized form (text, images, articles, audio files, etc)
Algorithm: Neural Network
Fully connected Neural Network, Convolutional Neural Network, Recurrent Neural Network, Transformer, etc

Depending on how you represent your problem, many algorithms can be used for both

**What are Neural Networks**

You have unstructured data as inputs. 

This data is then turned into numbers via numerical encoding so that the computer can understand them. The numbers are in matrix form (tensor)

Choose the appropriate neural netowrk for your problem, which learns representation (patterns/features/weights)

There is then representation outputs, also represnted as a weight tensor (learned representation). These can then be converted into human-understandable outputs



The "deep" in deep learning comes from having a lot of hidden layers

Overall architecture talks about the input layer, hidden layer, and output layer. 

Each layer is usually a combination of linear (straight line) and/or non-linear (non straight line) functions. The neural network is drawing patterns in the data

**Types of Learning**

Supervised Learning --> you have data and labels. You have a lot of examples of what the inputs should look like

Unsupervised and Self Supervised Learning --> You just have the data itself. You only have the photos you don't have the labels for cat and dog. You can get algorithm to learn inherent representation and fundamental patterns. It can detect "cat" and "dog" patterns but won't really know what is a cat and what is a dog

Transfer Learning --> Takes patterns that one model has learned and transfers to another model. The other model then gets a headstart

Reinforcement Learning --> you have an environment and an agent that does actions in that environment. You give reward and observations back to that agent

**What is Deep Learning Actually Used For**

Recommendation Page on Youtube
Google Translate, Speech Recognition --> seq2seq
Computer Vision, Natural Language Processing --> Classification and Regression

**What is PyTorch**

Most popular research deep learning framework

Write fast deep learning code in Python (able to run on a GPU/many GPUs)

Able to access many pre-built deep learning models (TorchHub/torchvision.models)

Whole stack: preprocess data, model data, deploy model in your application/cloud

Originally designed and used in-house by FB/Meta but is now open-source and used by Tesla, OpenAI, etc

PyTorch is a research favorite

Tesla is using PyTorch for computer vision frameworks for autopilot

OpenAI is standardized on PyTorch

GPU/TPU --> Graphics Processing Unit / Tensor Processing Unit

Pytorch enables you to leverage a GPU via CUDA interface (CUDA is a parallel computing platform and API that allows software to use certain types of GPUs for general purpose processing)
