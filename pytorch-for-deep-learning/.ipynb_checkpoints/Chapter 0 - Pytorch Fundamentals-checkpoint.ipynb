### Random Tensors

Why random tensors?

Random tensors are important because the way many NNs learn is that they start with tensors full of random numbers and then adjust those random numbers to better represent the data

Start with random nums --> look at data --> update random nums --> look at data --> update random numbers

```python
# create a random tensor of size (3, 4)

random_tensor = torch.rand(3,4)
random_tensor
```

```python
random_tensor.ndim
```

```python
random_tensor2 = torch.rand(1, 10, 10)
random_tensor2
```

```python
random_tensor3 = torch.rand(10, 10, 10)
random_tensor3
```

```python
# create a random tensor with a similar shape to an image tensor
random_image_size_tensor = torch.rand(size=(224, 224, 3)) # height, width, color channel
random_image_size_tensor.shape, random_image_size_tensor.ndim
```

### Matrix Multiplication

Two main ways of performing multiplication in neural network and deep learning
- element-wise multiplication
- matrix multiplication (dot product)

For matrix multiplication, the inner dimensions must match. Also, the resulting matrix has the shape of the outer dimensions (basic Lin Alg)

```python
print(tensor, "*", tensor)
print(tensor * tensor) # element-wise multiplication
```

```python
# Matrix Multiplication
torch.matmul(tensor, tensor) # 1 * 1 + 2 * 2 + 3 * 3
```

```python
%%time
value = 0
for i in range(len(tensor)):
    value += tensor[i] * tensor[i]
print(value)
```

```python
%%time
torch.matmul(tensor, tensor) # much faster than writing a for loop
```

### One of the most common errors in Deep Learning is Shape Errors

```python
torch.matmul(torch.rand(3,2), torch.rand(2,3))
```

```python
torch.matmul(torch.rand(2,3), torch.rand(3,2))
```

```python
tensor_a = torch.tensor([[1,2],
                        [3,4],
                        [5,6]])

tensor_b = torch.tensor([[7,10],
                        [8,11],
                        [9,12]])
torch.mm(tensor_a, tensor_b) # torch.mm is same as torch.matmul
```

To fix the tensor shape issues we can manipulate the shape of one of our tensors using the transpose. Transpose switches axes or dimensions of a given tensor


```python
tensor_b
```

```python
tensor_b.T, tensor_b.T.shape
```

```python
torch.matmul(tensor_a, tensor_b.T)
```

### Reshaping, Stacking, Squeezing, and Unsqueezing

* Reshaping - Reshapes an input tensor to a defined shape
* View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking - combine multiple tensors on top of each other (vstack) or side by side
* Squeeze - Removes all `1` dimensions from a tensor
* Unsqueeze - Add a `1` dimension to a target tensor
* Permute - Return a view of the input with dimensions permuted (swapped) in a certain way

```python
x = torch.arange(1., 10.)
x, x.shape
```

```python
# add an extra dimension
x_reshaped = x.reshape(1, 7) # trying to squeeze tensor of 9 elements into 7
```

```python
x_reshaped = x.reshape(1, 9)
```

```python
x_reshaped = x.reshape(2, 9) # cannot double the amount of elements
```

```python
x_reshaped = x.reshape(9, 1)
x_reshaped, x_reshaped.shape # reshape must be compatible with the original size
```

```python
x_reshaped = x.reshape(3, 3)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1, 9)
z, z.shape
```

```python
# changing z changes x because a view of a tensor shares the same memory as the original tensor
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked, x_stacked.shape
```

```python
x_stacked = torch.stack([x, x, x, x], dim=1)
x_stacked, x_stacked.shape
```

```python
x_v_stacked = torch.vstack([x,x,x,x])
x_v_stacked, x_v_stacked.shape
```

```python
x_h_stacked = torch.hstack([x,x,x,x])
x_h_stacked, x_h_stacked.shape
```

```python
y = torch.randn(2,3)
y, y.shape
```

```python
y_unsqueezed0 = torch.unsqueeze(y, dim=0)
y_unsqueezed0, y_unsqueezed0.shape
```

```python
y_squeezed_dim1 = torch.squeeze(y, dim=1)
y_squeezed_dim1, y_squeezed_dim1.shape # dimsion 1 is squeezed
```
